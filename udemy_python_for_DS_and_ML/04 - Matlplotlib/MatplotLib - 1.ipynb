# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
print(x)
print(y)
```

```python
# a simple 2-d plot
plt.plot(x,y)
```

```python
# Adding x,y labels and title to the plot
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# subplot: To draw multiple plots on the same canvas

# A canvas of 2x2 plots
# Display #1 and #4 plots in the canvas

plt.subplot(2,2,1)   # subplot(rows, cols, figure to work on)
plt.plot(x,y)

plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Object oriented method for plotting

OO methods give more control over the plots. Lets see how.

1. Get a figure object with `plt.figure()`.
2. Add axes to the figure
3. On the axes, plot the graph, add lables, add title etc.

```python
fig = plt.figure()   # A figure object
```

```python
fig.add_axes([0.1, 0.1, 0.8, 0.8])
```

```python
# All these commands must be executed together for the plot to be displayed
fig = plt.figure() 
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
# The dimensions in add_axes are [left, bottom, width, height] of the new axes. All
#    quantities are in fractions of figure width and height.
    
fig = plt.figure()

fig.add_axes([0.1, 0.1, 0.8, 0.8])
fig.add_axes([0.2, 0.4, 0.8, 0.8])
```

```python
fig = plt.figure()

# 2 axes with different width

fig.add_axes([0.1, 0.1, 0.8, 0.8])
fig.add_axes([0.1, 0.1, 0.5, 0.8])
```

```python
fig = plt.figure()

# 2 axes with different height

fig.add_axes([0.1, 0.1, 0.8, 0.8])
fig.add_axes([0.1, 0.1, 0.8, 0.5])
```

```python
fig = plt.figure()

fig.add_axes([0, 0, 1, 1])
fig.add_axes([0.1, 0.1, 0.8, 0.8])
```

```python
fig = plt.figure()

fig.add_axes([0, 0, 1, 1])
fig.add_axes([0.4, 0, 0.6, 0.6])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_xlabel('X1')

axes2.plot(y,x)
axes2.set_xlabel('X2')
```

> Use `plt.subplots()` to add multiple plots in the same figure.

It is a more succint way to plot multiple figures.

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# nrows and ncols are optional

fig, axes = plt.subplots(2, 1, figsize=(12, 5))
```

```python
fig, axes = plt.subplots(1,2)
for curr_ax in axes:
    curr_ax.plot(x,y)

plt.tight_layout()  # Recommendation is to put this at the end of you plot statements
```
