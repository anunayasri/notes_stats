# MatpltoLib - 2

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
print(x)
print(y)
```

## Multiple Graphs on Same Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2)
ax.plot(x, x**3)

# Plots multiple graphs on the same axis
# Auto colors different curves with different colors
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# Add legends

ax.plot(x, x**2, label='Square fn')
ax.plot(x, x**3, label='Cube fn')

ax.legend()  # Adds legend based on the 'label' property of each plot. 
             # You can also change the location of the legend(see docs - https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html)
```

## Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='red')  # or RGB hex codes '#FFEE02'
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# LineWidth and Transparency(alpha)

ax.plot(x,y, color='red', linewidth=4, alpha=0.2) 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# Line style

ax.plot(x,y, color='red', linestyle='-.')  # Multiple linestyle options available
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# markers represent the data-points on the line. This can be really helpful !

ax.plot(x,y, color='blue', lw=3, ls='-', marker='o', markersize=10,
       markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='green')
```
